```python
# notebooks/eda.ipynb

# 1. Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración general
pd.set_option('display.max_columns', 100)
sns.set(style="whitegrid")
%matplotlib inline

# 2. Carga de datos
df = pd.read_csv("../data/train.csv")
print("Shape del dataset:", df.shape)
df.head()

```

```python
# 3. Información general
df.info()
df.describe()
```

```python
# 4. Valores nulos
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print("Columnas con valores nulos:\n", missing)
```

```python
# 5. Visualización básica: Precio de venta
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], kde=True)
plt.title("Distribución del precio de venta")
```

```python
# 6. Correlación con SalePrice
correlation = df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
print("Variables más correlacionadas con SalePrice:\n", correlation.head(10))
```

```python
# 7. Heatmap de correlaciones principales
top_corr = correlation.index[:10]
plt.figure(figsize=(10, 8))
sns.heatmap(df[top_corr].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap de variables más correlacionadas con el precio")
```

```python
# 8. Análisis de algunas variables clave
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title("Precio según calidad general de la casa")
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title("Área habitable vs Precio de venta")
```

```python
# 9. Scatterplot para detectar outliers en GrLivArea
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title("Área habitable (GrLivArea) vs Precio")
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Neighborhood', y='SalePrice', data=df)
plt.xticks(rotation=45)
plt.title("Precio según vecindario")
plt.figure(figsize=(10, 5))
sns.boxplot(x='HouseStyle', y='SalePrice', data=df)
plt.xticks(rotation=45)
plt.title("Precio según estilo de casa")
```

```python
# 10 Ver qué tipos de variables tienen más nulos
missing_summary = df.isnull().sum().reset_index()
missing_summary.columns = ['column', 'missing']
missing_summary = missing_summary[missing_summary['missing'] > 0]
missing_summary['dtype'] = missing_summary['column'].apply(lambda x: df[x].dtype)
missing_summary.sort_values(by='missing', ascending=False)
```
